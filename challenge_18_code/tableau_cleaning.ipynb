```python
just_end = summer_2024_df[["end_station_name", "end_lat", "end_lng"]]
just_end.head(20)
```

```python
# Create a dictionary to map station names to their correct coordinates from the end station columns
station_coords = summer_2024_df[['end_station_name', 'end_lat', 'end_lng']].drop_duplicates(subset='end_station_name').set_index('end_station_name').to_dict(orient='index')

# Function to update the start coordinates based on the correct end coordinates
def update_start_coords(row):
    station = row['start_station_name']
    if station in station_coords:
        row['start_lat'] = station_coords[station]['end_lat']
        row['start_lng'] = station_coords[station]['end_lng']
    return row

# Apply the function to the dataframe to update start coordinates
summer_2024_df = summer_2024_df.apply(update_start_coords, axis=1)
```
